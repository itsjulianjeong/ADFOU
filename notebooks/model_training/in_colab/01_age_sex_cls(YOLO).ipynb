# 모델 학습
현재까지 진행사항
1. Training 폴더의 img 파일을 읽어와서 json 파일을 기준으로 전처리
    - 나이/성별 csv 파일 생성(labels.csv)
    - 얼굴 crop 후 저장(로컬 cropped_faces 폴더)
2. cropped_faces 폴더 불러와서
    - 성별 + 나이 클래스 조합 -> 클래스명으로 생성(class_name) -> 클래스명 -> 고유 인덱스 매핑
    - stratify를 위해 class_name 컬럼 기준으로 분리(YOLO_cls_dataset -> train/val 0.2로)
    - 해당 경로로 이미지 복사해옴
3. 로컬에서 작업완료  
    - 코랩으로 불러온 파일들: YOLO_cls_dataset.zip(labels.csv, YOLO_cls_dataset)

## 모델 학습

````python
from ultralytics import YOLO

model = YOLO("yolov8s-cls.pt")
model.train(
    data="/content/drive/MyDrive/Colab Notebooks/vision_project/age_sex_cls_project/YOLO_cls_dataset",
    imgsz=224,
    epochs=10,
    batch=32,
    patience=3,
    name="age_sex_cls_colab"
)
````
